```python
counter = {}
files = []
with open("./calib_splits/am_imagenet_5percent_val.txt", "r") as f:
    for line in f:
        target = line.split()[1]
        img_file = line.split()[0]
        files.append(img_file)
        
        if target not in counter:
            counter[target] = 1
        else:
            counter[target] += 1
    
```

```python
train_counter = {}
train_files = []
with open("./calib_splits/am_imagenet_5percent_train.txt", "r") as f:
    for line in f:
        target = line.split()[1]
        img_file = line.split()[0]
        train_files.append(img_file)
        
        if target not in train_counter:
            train_counter[target] = 1
        else:
            train_counter[target] += 1
```
